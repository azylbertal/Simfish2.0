```python
import numpy as np
import h5py

predator_loc = np.zeros((1, 1000, 2))

for ii, pred_triplet in enumerate([200, 300, 400, 500, 600, 700]):
    if ii % 2 == 0:
        origin = 2600
    else:
        origin = 2400
    predator_loc[0, pred_triplet:pred_triplet+9, 0] = np.linspace(origin, 2500, 9)
    predator_loc[0, pred_triplet:pred_triplet+9, 1] = 2550
    predator_loc[0, pred_triplet+10:pred_triplet+19, 0] = np.linspace(origin, 2500, 9)
    predator_loc[0, pred_triplet+10:pred_triplet+19, 1] = 2550
    predator_loc[0, pred_triplet+20:pred_triplet+29, 0] = np.linspace(origin, 2500, 9)
    predator_loc[0, pred_triplet+20:pred_triplet+29, 1] = 2550

prey_loc = np.zeros((2, 1000, 2))

for ii, prey_init in enumerate([230, 330, 430, 530, 630, 730]):
    prey_loc[0, prey_init:prey_init+20, 0] = np.linspace(2400, 2500, 20)
    prey_loc[0, prey_init:prey_init+20, 1] = 2530
    prey_loc[1, prey_init:prey_init+20, 0] = np.linspace(2600, 2500, 20)
    prey_loc[1, prey_init:prey_init+20, 1] = 2530

with h5py.File('prey_stim.h5', 'w') as f:
    f.create_dataset('prey_loc', data=prey_loc)

with h5py.File('predator_stim.h5', 'w') as f:
    f.create_dataset('predator_loc', data=predator_loc)
```
